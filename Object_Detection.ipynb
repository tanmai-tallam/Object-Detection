```python
# haar-cascade algorithm for eye detection
```

```python
img=cv2.imread("/content/p1.png")
cv2_imshow(img)
```

```python
# read the image by pasting the path
```

```python
img=cv2.imread("/content/p1.png")
eyes=cv2.CascadeClassifier("/content/haarcascade_eye.xml")
e=eyes.detectMultiScale(img,1.1,10)
for (x,y,w,h) in e:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2_imshow(img)
```

```python
# Detection of eyes is detected by red square boxes as shown in the above by using the above code
```

```python
b=cv2.imread("/content/p3.png")
numberplate=cv2.CascadeClassifier("/content/haarcascade_licence_plate_rus_16stages.xml")
n=numberplate.detectMultiScale(b,1.02,5)
for (x,y,w,h) in n:
    cv2.rectangle(b,(x,y),(x+w,y+h),(0,0,255),2)
cv2_imshow(b)
```
